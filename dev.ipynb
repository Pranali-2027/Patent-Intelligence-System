```python
import os
import json
import requests
from dotenv import load_dotenv
```

```python
load_dotenv()

api_key = os.getenv("SERP_API_KEY")

if not api_key:
    raise ValueError("SERP_API_KEY is not set.")
```

```python
query = "lithium battery"

url = f"https://serpapi.com/search.json?engine=google_patents&q={query}&api_key={api_key}"

print(url)
```

```python
response = requests.get(url, timeout=20)
print(response.status_code)

data = response.json()
print(data.keys())
```

```python
patents = data.get("organic_results", [])
print(len(patents))
```

```python
os.makedirs("files", exist_ok=True)
```

```python
with open("files/patents.json", "w") as file:
    json.dump(patents, file, indent=4)

print("Patent data saved successfully.")
```

```python
for patent in patents[:5]:
    print("Title:", patent.get("title"))
    print("Inventor:", patent.get("inventor"))
    print("Publication:", patent.get("publication_number"))
    print("Link:", patent.get("patent_link"))
    print("-" * 50)
```

```python
patent_url = "https://serpapi.com/search.json?engine=google_patents_details&patent_id=patent%2FCN119480895B%2Fen" + api_key
patents_response = requests.get(patent_url)
if patents_response.status_code == 200:
    patent_data = patents_response.json()

    with open("files/patent_details.json", "w") as file:
        json.dump(patent_data, file, indent=4)
```

```python
dir_files = os.listdir("files")

for file in dir_files:
    if file.endswith(".json"):

        with open(os.path.join("files", file), "r") as f:
            content = json.load(f)

            for patent in content[:5]:

                print("Title:", patent.get("title"))
                print("Inventor:", patent.get("inventor"))
                print("Assignee:", patent.get("assignee"))
                print("Publication:", patent.get("publication_number"))
                print("Patent Link:", patent.get("patent_link"))

                print("\n" + "=" * 50 + "\n")
```
